Due to data protection for the users, and because we dont need these columns, they are dropped

```python
df.drop(columns=["date", "name", "user"], axis = 1, inplace = True)
df.head()
```

```python
df.drop_duplicates(inplace = True)

df.shape
```

Once duplicates and null values are removed, we lose roughly 100,000 rows, however this is neccesary to ensure the most accurate ranking.

```python
df.dropna(inplace = True)

df.shape
```
