In order to match the results of the language models we must add an abbreviation column to the actual dataset of rankings. The universities used in the named entity recognition part of the project are the top 50 ranked universities in the actual list. These university names are often large and or formatted in a way that the NER system will not recognise, as such we manually assign abbreviations to each university and write it back into the csv.
