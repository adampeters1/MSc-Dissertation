```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertForSequenceClassification

# Load the pre-trained BERT model and tokenizer
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)
model.eval()  # Set the model to evaluation mode


def analyze_sentiment(text):
    # Tokenize the text and convert it to input features
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)

    # Make a forward pass through the model
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the predicted sentiment score (logits)
    logits = outputs.logits

    # Convert logits to probabilities using softmax
    probabilities = torch.softmax(logits, dim=1)

    # Return the probabilities for positive and negative sentiments
    return probabilities[0][1].item()  # Probability of being positive
```

```python
df['bert_base_uncased_score'] = df['content'].apply(analyze_sentiment)

df = df.sort_values(by='bert_base_uncased_score', ascending=False)
df.head(50)

from google.colab import drive

drive.mount('/content/gdrive')
path = '/content/gdrive/My Drive/Dissertation Complete/BERT_base_uncased_results.csv'

with open(path, 'a') as f:
  df.to_csv(f, encoding='utf-8', index=False, header=False)
```
